```python
import random
import os
from PIL import Image, ImageDraw, ImageFont

def get_random_font():
    # Get the path to the "fonts" directory
    fonts_directory = "fonts"

    # Get a list of all font files in the "fonts" directory
    font_files = [os.path.join(fonts_directory, file) for file in os.listdir(fonts_directory) if file.endswith((".ttf", ".otf"))]

    if not font_files:
        return None

    return random.choice(font_files)

def generate_quote_image(quote, author):
    # Define the image size
    width, height = 1080, 1350 

    # Randomly pick a background color
    background_color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

    # Create a new image with the randomly selected background color
    image = Image.new("RGB", (width, height), background_color)
    draw = ImageDraw.Draw(image)

    # Determine a suitable text color based on background luminance
    luminance = (0.299 * background_color[0] + 0.587 * background_color[1] + 0.114 * background_color[2]) / 255
    text_color = (0, 0, 0) if luminance > 0.5 else (255, 255, 255)

    # Choose a random font size for the quote and author
    quote_font_size = random.randint(100, 120)
    author_font_size = random.randint(65, 75)

    # Randomly pick a font for the quote and author
    quote_font_file = get_random_font()
    author_font_file = quote_font_file

    # Load the selected fonts with the specified sizes
    try:
        quote_font = ImageFont.truetype(quote_font_file, quote_font_size)
        author_font = ImageFont.truetype(author_font_file, author_font_size)
    except (IOError, TypeError):
        # Use default font if the random font is not available
        quote_font = ImageFont.truetype("arial.ttf", quote_font_size)
        author_font = ImageFont.truetype("arial.ttf", author_font_size)

    # Calculate the maximum allowed width for the quote and author texts
    max_text_width = width - 100

    # Fit the quote text within the image width
    quote_lines = []
    current_line = ""
    words = quote.split()
    for word in words:
        if quote_font.getsize(current_line + word)[0] <= max_text_width:
            current_line += word + " "
        else:
            quote_lines.append(current_line.strip())
            current_line = word + " "
    quote_lines.append(current_line.strip())

    # Calculate the position to center the quote text vertically
    total_quote_height = sum(quote_font.getsize(line)[1] for line in quote_lines)
    y_quote = (height - total_quote_height) // 2

    # Draw the quote lines on the image
    for line in quote_lines:
        line_width, line_height = draw.textsize(line, font=quote_font)
        x_quote = (width - line_width) // 2
        draw.text((x_quote, y_quote), line, fill=text_color, font=quote_font)
        y_quote += line_height

    # Calculate the position for the author text
    author_width, author_height = draw.textsize(author, font=author_font)
    x_author = (width - author_width) // 2
    y_author = y_quote + 20  # Add some spacing below the quote

    # Draw the author text just below the quote
    draw.text((x_author, y_author), "- "+author, fill=text_color, font=author_font)

    return image

# Example usage:
quote_text = "If you're going through hell, keep going."
author_text = "Winston Churchill"
image = generate_quote_image(quote_text, author_text)
image.show()  # To display the image using the default image viewer
```

```python
import gradio as gr

# Gradio Interface
inputs = [
    gr.inputs.Textbox(lines=3, label="Quote"),
    gr.inputs.Textbox(label="Author")
]

output = gr.outputs.Image(type='pil')

gr.Interface(fn=generate_quote_image, inputs=inputs, outputs=output, title="Quote Image Generator").launch()
```
